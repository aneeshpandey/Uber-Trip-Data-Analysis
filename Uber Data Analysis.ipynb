```python
%pylab inline
import pandas
import seaborn
```

# Load CSV into memory

```python
data=pandas.read_csv('C:/Users/Aneesh Pandey/Desktop/Programming/Uber Data Analysis/uber-raw-data-apr14.csv')
```

```python
data.head()
```

# Convert to datetime and add some useful columns


```python
dt=data['Date/Time'][1]
dt
```

```python
dt=pandas.to_datetime(dt)
```

```python
data['Date/Time']=data['Date/Time'].map(pandas.to_datetime)
```

```python
data.tail()
```

```python
def get_dom(dt):
    return dt.day

data['dom']=data['Date/Time'].map(get_dom)
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday']=data['Date/Time'].map(get_weekday)
```

```python
def get_hour(dt):
    return dt.hour

data['hour']=data['Date/Time'].map(get_hour)
```

```python
data.tail()
```

# Analysis

## Analyze the DoM

```python
hist(data.dom,bins=30,rwidth=0.8,range=(0.5,30.5))
xlabel('Date of the month')
ylabel('Frequency')
title('Frequency by DoM - Uber - Apr 2014')
```

```python
#for k,rows in data.groupby('dom'):
#    print(k,len(rows))

def count_rows(rows):
    return len(rows)

by_date=data.groupby('dom').apply(count_rows)
by_date
```

```python
plot(by_date)
```

```python
bar(range(1,31),by_date);
```

```python
sorted_by_date=by_date.sort_values()
sorted_by_date
```

```python
bar(range(1,31),sorted_by_date,width=0.7);
xticks(range(1,31),sorted_by_date.index);
xlabel('Date of the month')
ylabel('Frequency')
title('Frequency by DoM - Uber - Apr 2014')
```

## Analyzing the hours

```python
hist(data.hour, bins=24,rwidth=0.8,range=(-0.5,23.5))
```

# Analyzing the Weekdays

```python
for k,rows in data.groupby('weekday'):
    print(k,len(rows))
```

```python
hist(data.weekday,bins=7,rwidth=0.8,range=(-0.5,6.5),color='green',alpha=0.6)
xticks(range(0,7),'Mon Tue Wed Thu Fri Sat Sun'.split(' '));
```

# Cross Analysis (hour,weekday)

```python
by_weekday_hour=data.groupby('weekday hour'.split(' ')).apply(count_rows).unstack()
by_weekday_hour
```

```python
seaborn.heatmap(by_weekday_hour)
```
